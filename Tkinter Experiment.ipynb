```python
from tkinter import *

class Controller:
    def kirim(self,e):
        value = e.get()
        print(value)
        

class View():
    name = "View"
    def __init__(self,top):
        self.e = Entry(top)
        self.Lb1 = Listbox(top)
        self.Lb1.insert(1, "Python")
        self.Lb1.insert(2, "Perl")
        self.Lb1.insert(3, "C")
        self.Lb1.insert(4, "PHP")
        self.Lb1.insert(5, "JSP")
        self.Lb1.insert(6, "Ruby")
        self.Lb1.bind('<<ListboxSelect>>', self.onselect)
        self.btn = Button(top, text="Kirim", command= lambda: Controller().kirim(self.e))
        self.e.pack()
        self.Lb1.pack()
        self.btn.pack()
        
    def onselect(self,evt):
        w = evt.widget
        print(int(w.curselection()[0]))
        index = int(w.curselection()[0])
        value = w.get(index)[0]
        self.e.delete(0,END)
        self.e.insert(END, value)

    def kirim(self):
        value = e.get()
        print(value)

    
top = Tk()
display = View(top)
top.mainloop()
```

```python
from tkinter import *

class App:
    def __init__(self,master):
        sidebar = Frame(master)
        sidebar.pack(side=LEFT, fill=BOTH)
        sidebar.configure(background="red")
        container = Frame(master)
        container.pack(side=LEFT, fill=BOTH, expand=1)
        container.configure(background="blue")
        btn_act = Button(sidebar, text="Activiy", width=20, command=lambda: self.activity(container))
        btn_lib = Button(sidebar, text="Library", width=20, command=lambda: self.library(container))
        btn_act.grid(row=0, column=0)
        btn_lib.grid(row=1, column=0)

    def activity(self,sidebar):
        container = sidebar
        container.pack(side=LEFT, fill=BOTH, expand=1)
        container.configure(background="blue")
        


    def library(self,sidebar):
        container = sidebar
        container.pack(side=LEFT, fill=BOTH, expand=1)
        container.configure(background="yellow")


master = Tk()
display = App(master)
master.geometry("500x500")
master.mainloop()
```
